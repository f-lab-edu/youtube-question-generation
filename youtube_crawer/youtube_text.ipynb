```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
# https://www.youtube.com/watch?v=BCkz6iR3xPU
srt = YouTubeTranscriptApi.get_transcript("BCkz6iR3xPU", languages=['ko'])

with open("subtitles.txt", "w", encoding="utf-8") as f:
  text = ''
  for i in srt:
    words = i['text'].split(' ')
    lst = ['니다', '세요', '고요', '구요', '데요', '까요', '에요', '어요', '내요','되요', '든요', '아요', '해요', '게요', '네요', '뭡니까']
    
    # 문장부호 붙이기
    sen = ''
    for word in words:
      for j in lst:
        if j == word[-len(j):]:
          word += '.'
      sen += word + ' '
      
    f.write("{}\n".format(sen))
```
